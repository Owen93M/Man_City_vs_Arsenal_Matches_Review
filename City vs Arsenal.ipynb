# Plots

## Bar

```python
# Data
home_goal_data = {'Manchester City': [6], 
           'Arsenal': [4]}

# Bar Colors
colors = ['skyblue', 'maroon']

# Create Df
df = pd.DataFrame(home_goal_data, columns=['Manchester City', 'Arsenal'], index=[''])

# Create a figure and axis with specific size
fig, ax = plt.subplots(figsize=(10, 6))  # Width = 10, Height = 6

# Plot the bar chart on the specified axes
df.plot.bar(color=colors, ax=ax)

# Plot Title And Y-Label
ax.set_title('Home Goals Scored', fontweight='bold')
ax.set_ylabel('Goals', fontweight='bold')

# Ensure the ticks are placed correctly
ax.set_xticks([0])  
ax.set_xticklabels([''])
ax.tick_params(axis='x', rotation=0)

# Set Background Color
ax.set_facecolor('#59656d')

# Plot Legend
ax.legend()

# Display Figure
plt.show()
```

```python
# Make a data definition
date = ['Arsenal v Manchester City\n (1-3)', 'Manchester City v Arsenal\n (4-1)', 'Arsenal v Manchester City\n (1-0)', 'Manchester City v Arsenal\n (0-0)', 'Manchester City v Arsenal\n (2-2)']
city = [9, 14, 12, 12, 33]
asnl = [9, 8, 4, 6, 11]

# np.arange = Function That Generates Arrays w/ Evenly Spaced Values Withing Defined Interval
x_axis = np.arange(len(date))
 
# Multi Bar Chart
# Plot Bar( x-axis *-0.2 = Left Of Bar Center, +0.2 = Right Of Bar Center*, label, color)
plt.bar(x_axis -0.2, city, width=0.5, label = 'City Shot', color='skyblue')
plt.bar(x_axis +0.2, asnl, width=0.5, label = 'Arsenal Shot', color='maroon')
 
# Xticks
# Use x-axis Variable & date Variable 
plt.xticks(x_axis, date)
 
# Add Labels & Style
plt.ylabel("#No. Shots", fontsize=11, fontweight='bold')
plt.xticks(rotation=45, ha='right')

# Plot Background Color With gca And Set Facecolor With Hex Number
plt.gca().set_facecolor('#6b7c85')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

## Pie Charts

#### Tackles 

```python
# Create Variable To Store Fixtures
fixtures = ['Arsenal v Manchester City','Manchester City v Arsenal','Arsenal v Manchester City','Manchester City v Arsenal','Manchester City v Arsenal']
    
# Create Varaible To Store Values
values = [13,11,20,13,10]

# Create Pie Chart
# Plot Figure - Set Figure Size 'figsize'
plt.figure(figsize=(10, 8))
# Call :
#  Wedges = Manipulate Wedge (explode)
#  Texts = Modify Outside Labels (size, style, color)
#  Autotexts = Modify Inside Labels (size, style, color)
# Plot Pie (Add Arguments/Parameters)
wedges, texts, autotexts = plt.pie(
# Select 'values' Varaible To Use Data To Show On SLices
    values,
# Create 'labels' Variable And Call 'fixtures' Variable To Use Data As Labels For Slices
    labels=fixtures,
# Select Distance Selected Slices Are Split From Others
    explode=[0.05,0,0.09,0,0],
# Creates & Shows Percentage Value Using 'values' Variable Data
    autopct=lambda p: f'{p * sum(values) / 100:.2f}',  
# Starting Point For The Pie Chart To Begin
    startangle=90,
# Styles Text Within Pie Chart
    textprops={'fontsize': 12, 'weight': 'bold'}
)

# Plot Title & Show
plt.title("Tackles By Home Team", fontsize=14, fontweight='bold')

# Ajust Spacing In Graph
plt.tight_layout()

# Show Open Figures
plt.show()
```
